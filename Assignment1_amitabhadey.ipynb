<p>The aspect ratio of the image is 1.1:1</p>

<p>Therefore, if the height is 261 pixels, to maintain the aspect ratio, 
the width of the image must be (243/220) * 261 = 288.3 pixels to be exact.

```python
# Define the window size
windowsize_r = 9
windowsize_c = 9

# Crop out the window and calculate the histogram
for r in range(0,resized_bana.shape[0] - windowsize_r, windowsize_r):
    for c in range(0,resized_bana.shape[1] - windowsize_c, windowsize_c):
        window_bana = resized_bana[r:r+windowsize_r,c:c+windowsize_c]
```

<h2>Task 11: Conceptual Question</h2>
    <p>
    <ul>
    <li>Describe the effects of block size on the dimensionality of the feature space and the number of vectors 
in the domain</li>
<li>Describe how these effect may influence the classifier that divides the domain</li>
</ul>
</p>

<b>Answer:</b>

<p>If we increase the block size, we will introduce more number of features. Datasets having more number of features 
than observations are known as high-dimensional data. Because these data have more dimensions, it increases 
complexity and reduces the effectiveness of computational and statistical techniques. 
A dataset having many dimensions doesn't necessarily ensure that all the features provide
useful information. This creates a problem known as dimensional redundancy. Therefore, we have to be careful in
choosing an optimal block size.<br>

If the size of the block size is less, it will have shift more over the original image to iterate through all the pixels.
As a result, the vector would be greater in that instance. On the other hand, if the size of the block is bigger, it 
will have to travel less distance to map the entire image, and therefore the vector would be lower in comparison.<br>

The classifier depends on the features. If the classifer has to take into consideration more number of features, it
is likely to be less effective because it needs to factor in large amount of information. Also, if there are 
certain features that do not help in the distinction between two classes, it will inversely impact the performance
of the classifier.</p>
